```python
import matplotlib.pyplot as plt #plotting device
import numpy as np #works with assigning range of numbers
from sympy import * #works with functions
PI = np.pi

x = Symbol('x')
y = lambdify(x, "x**2+3")
h = lambdify(x, "5-x**2")

t = np.linspace(0,1)
plt.plot (t,y(t))
plt.plot (t,h(t))
plt.show()
```

```python
integral = integrate(PI*((5-x**2)**2 - (x**2+3)**2), (x, 0, 1))
print (integral)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection='3d')

u = np.linspace(0, 2, 60)
v = np.linspace(0, 2*np.pi, 60)
U, V = np.meshgrid(u, v)

X = U
Y = (U**2 + 3)*np.cos(V)
Z1 = (U**2 + 3)*np.sin(V)

H = (5-U**2)*np.cos(V)
Z2 = (5-U**2)*np.sin(V)

ax.plot_surface(X, Y, Z1, color='yellow')
ax.plot_surface(X, H, Z2, color='blue')
plt.show()
```
